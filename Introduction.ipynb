## torch.unsqueeze(tensor, dim)
#### Inserts a new dimension of size 1 at 'dim'. For example, it can change a 1-d tensor to a 2-d tensor (effectively unsqueezing the tensor), or change a 2-d tensor to a 3-d tensor.

```python
x = torch.Tensor(2,3)
print(x.size())
print("2-d tensor x: ", x)
temp = torch.unsqueeze(x, 0)
print(temp.size())  # <- 3-d tensor.
print(f"Unsqueezed x (changes to 3-d at dimension 0): ", temp)
```

## torch.squeeze(tensor, dim=None)
#### Removes all dimensions of size 1 from the tensor.
